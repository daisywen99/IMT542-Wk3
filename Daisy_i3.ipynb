```python
# Install necessary libraries (only need to run once)
!pip install pandas matplotlib seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#from google.colab import files
uploaded = files.upload()

import io
movies = pd.read_csv(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

movies.head()


# Quick look at data
movies.head()

# Check basic info
print("Number of Movies:", len(movies))
print("\nColumns Available:", movies.columns.tolist())

# --- Visualization 1: Top 10 Highest Rated Movies ---
top_movies = movies.sort_values(by='IMDb Rating', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='IMDb Rating', y='Title', data=top_movies)
plt.title('Top 10 Highest Rated Movies')
plt.xlabel('IMDb Rating')
plt.ylabel('Movie Title')
plt.show()


# --- Visualization 2: Number of Movies by Genre ---
plt.figure(figsize=(12,6))
movies['Genres'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Genres by Number of Movies')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()


# --- Visualization 3: Rating Distribution ---
plt.figure(figsize=(10,5))
sns.histplot(movies['IMDb Rating'], bins=20, kde=True)
plt.title('Distribution of IMDb Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Frequency')
plt.show()

```
